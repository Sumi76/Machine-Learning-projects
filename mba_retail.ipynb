```python
import pandas as pd
df=pd.read_csv('Analytics in Retail Project/Dataset/Groceries_dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df['itemDescription'].unique()
```

```python
basket=df.groupby(['Member_number','itemDescription'])['itemDescription'].count().unstack().fillna(0)
basket
```

```python
def encode(x):
    if x>0:
        return 1
    if x<=0:
        return 0
```

```python
encoded_b=basket.applymap(encode)
encoded_b
```

```python
encoded_b['Instant food products'].value_counts()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
apr=apriori(encoded_b,min_support=0.07,use_colnames=True)
rules=association_rules(apr,metric='lift',min_threshold=1)
#rules=rules.sort_values(['confidence'],ascending=False)
rules
```

```python
def get_recommendations(product,rules,num_recommendations=5):
    recommendations=[]
    for idx,row in rules.iterrows():
        if product in row['antecedents']:
            recommended_item=next(iter(row['consequents']))
            recommendations.append((recommended_item,row['lift']))
    recommendations.sort(key=lambda x: x[1], reverse=True)    
    return recommendations[:num_recommendations]  
```

```python
product_to_recommend='yogurt'
recommendations=get_recommendations(product_to_recommend,rules)
print(f"Recommendations for'{product_to_recommend}':")
for item, lift in recommendations:
    print(f"{item}(lift:{lift:.2f})")
```
