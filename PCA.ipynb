```python
import pandas as pd
df=pd.read_csv('/content/telecom_churn.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df1=df.drop('phone number',axis=1)
df1
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
label=preprocessing.LabelEncoder()
df1['state']=label.fit_transform(df1['state'])
df1['international plan']=label.fit_transform(df1['international plan'])
df1['voice mail plan']=label.fit_transform(df1['voice mail plan'])
df1['churn']=label.fit_transform(df1['churn'])
df1.info()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
x=df1
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

wcss=[]
for i in range(2,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x_scaled)
  wcss.append(kmeans.inertia_)

  #plot elbow graph
plt.figure(figsize=(8,6))
plt.plot(range(2,11),wcss,marker='o',linestyle="--")
plt.title('elbow mwthod')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=56)
kmeans.fit(x_scaled)
df['Cluster']=kmeans.labels_
df
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
pca=PCA(n_components=1)
dfs=pca.fit_transform(df1)
kmeans=KMeans(n_clusters=3,random_state=100)
kmeans.fit(dfs)
label=kmeans.labels_
print(silhouette_score(dfs,label))
```

```python
pca=PCA(n_components=2)
dfs=pca.fit_transform(df1)
kmeans=KMeans(n_clusters=3,random_state=100)
kmeans.fit(dfs)
label=kmeans.labels_
print(silhouette_score(dfs,label))
```

```python
pca=PCA(n_components=3)
dfs=pca.fit_transform(df1)
kmeans=KMeans(n_clusters=3)
kmeans.fit(dfs)
label=kmeans.labels_
print(silhouette_score(dfs,label))
```

```python
pca=PCA(n_components=4)
dfs=pca.fit_transform(df1)
kmeans=KMeans(n_clusters=3)
kmeans.fit(dfs)
label=kmeans.labels_
print(silhouette_score(dfs,label))
```

```python
sil_score=[]
for k in range(2,11):
  kmeans=KMeans(n_clusters=k,random_state=42)
  kmeans.fit(dfs)
  labels=kmeans.predict(dfs)
  sil_score.append(silhouette_score(dfs,labels))
plt.plot(range(2,11),sil_score)
```
