```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
iris=datasets.load_iris()
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
adamodel=AdaBoostClassifier(n_estimators=100,learning_rate=1)
model=adamodel.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier(random_state=100)
random_forest.fit(x_train,y_train)

y_pred_rf=random_forest.predict(x_test)
accuracy_rf=metrics.accuracy_score(y_test,y_pred_rf)
print('random forest accuracy:',accuracy_rf)
```

```python
df=df.drop(' thinness  1-19 years',axis=1)
df=df.drop('under-five deaths ',axis=1)
```

```python
import xgboost as xgb
xg_reg = xgb.XGBRegressor()
x=df.drop('Life expectancy',axis=1)
y=df['Life expectancy']
X=pd.get_dummies(df)
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
xg_reg.fit(X_train, y_train)
yxg_pred = xg_reg.predict(x_test)

mse = metrics.mean_squared_error(y_test, yxg_pred)
mae=metrics.mean_absolute_error(y_test,yxg_pred)
r2_xgb =metrics. r2_score(y_test, yxg_pred)

print("Mean Squared Error: ", mse)
print('Mean Absolute Error: ',mae)
print("R^2 Score: ", r2_xgb)




```
